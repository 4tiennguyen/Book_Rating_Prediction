### Univariate Selection

```python
from sklearn.decomposition import PCA
```

# am_gr_metadata

```python
# Load am_gr_metadata metadata
am_gr = pd.read_csv('am_gr_metadata.csv')
am_gr
```

# Goodreads Metadata

```python
# Load Goodreads metadata
gr = pd.read_csv('goodreads_metadata.csv')
gr
```

```python
gr = pd.merge(gr, am_gr[['asin','rating_diff']], how='inner', on='asin')
gr
```

### Univariate Selection

```python
array = gr.values
X = array[:,1:7]
Y = array[:,7]
# feature extraction
test = SelectKBest(score_func=f_classif, k=3)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
```

Univariate Selection best 3 features to predict average_rating: total_reviews_count, total_ratings_count, total_text_reviews_count

```python
array = gr.values
X = array[:,1:7]
Y = array[:,8]
# feature extraction
test = SelectKBest(score_func=f_classif, k=3)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
```

Univariate Selection best 3 features to predict rating_diff: total_reviews_count, total_text_reviews_count, reviews_count

### PCA

```python
array = gr.values
X = array[:,1:7]
Y = array[:,7]
# feature extraction
pca = PCA(n_components=3)
fit = pca.fit(X)
# summarize components
print("Explained Variance: %s" % fit.explained_variance_ratio_)
```
