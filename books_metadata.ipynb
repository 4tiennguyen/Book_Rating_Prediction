```python
import numpy as np
import pandas as pd
import json
```

```python
gr = pd.read_csv('cleaned_goodreads.csv')
```

```python
gr
```

```python
# Find IDs of unique books to find metadata for
books = np.unique(gr.book_id)
print(len(books))
```

```python
# Create empty dataframe to store books metadata
df = pd.DataFrame(columns = ['book_id', 'books_count', 'original_publication_month', 'original_publication_year', 
                             'rating_dist', 'genres'])
```

```python
# Add books & genres to dataframe
count = 0
with open('goodreads_book_genres_initial.json', 'r') as f:
    for line in f:
        if count == len(books):
            break
        else:
            j = json.loads(line)
            if int(j['book_id']) in books: 
                df = df.append({'book_id':j['book_id'], 'genres':j['genres']}, ignore_index=True)
                count += 1
```

```python
df
```

```python
df.to_csv('genres.csv')
```

```python
# Add metadata to dataframe
count = 0
with open('goodreads_book_works.json', 'r') as f:
    for line in f:
        if count == len(books):
            break
        else:
            j = json.loads(line)
            if int(j['work_id']) in books: 
                df.loc[df.book_id == j['work_id'],'books_count'] = j['books_count']
                df.loc[df.book_id == j['work_id'],'original_publication_month'] = j['original_publication_month']
                df.loc[df.book_id == j['work_id'],'original_publication_year'] = j['original_publication_year']
                df.loc[df.book_id == j['work_id'],'rating_dist'] = j['rating_dist']
                count += 1
```

```python
df
```

```python
df.to_csv('metadata.csv')
```
