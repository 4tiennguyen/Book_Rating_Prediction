```python
import numpy as np
import pandas as pd
import json
```

```python
import json
l = []
data =[]
with open('Books_5.json', 'r') as f:
    for i, line in enumerate(f):
        if i == 100000:
            break
        j = json.loads(line)
        if not 'reviewText' in j: 
            continue
        l.append([j['overall'], j['reviewTime'], j['asin'], j['reviewText']])
```

```python
data = pd.DataFrame(l, columns = ['overall', 'reviewTime', 'asin', 'reviewText'])
# year < 2013
data[data['reviewTime'].str[-4:].astype(int) < 2013]
```
