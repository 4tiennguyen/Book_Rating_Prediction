```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the first 2M reviews from Amazon dataset

```python
# read the first 2M reviews
l = []
data =[]
with open('Books_5.json', 'r') as f:
    for i, line in enumerate(f):
        if i == 2000000:
            break
        j = json.loads(line)
        if not 'reviewText' in j: 
            l.append([j['overall'], j['reviewTime'], j['asin'], np.nan ])
        else :
            l.append([j['overall'], j['reviewTime'], j['asin'], j['reviewText']])
    print(j.keys())
```

```python
data = pd.DataFrame(l, columns = ['overall', 'reviewTime', 'asin', 'reviewText'])
```

```python
data
```

```python
data.to_csv('amazonData_1.csv',index=False)
```

# Spliting file into smaller files (only asin)

```python
import os

n = 29718
N = 10
k = 0

n_per_file = int(n // N) + 1

with open('asinOnly1.csv', 'r') as f:
    next(f)
    g = None
    for i, asin in enumerate(f):
        if i % n_per_file == 0:
            k = k + 1
            if not g == None:
                g.close()
            g = open('asinOnly{}.csv'.format(k), 'w')
            g.write("asin\n")
        g.write("{}".format(asin))
    g.close()
```
